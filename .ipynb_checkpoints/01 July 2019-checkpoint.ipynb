```python
import pandas as pd
filepath='DataFiles/Income.csv'
def readDf(filepath):
    incomedf = pd.read_csv(filepath)
    return incomedf
readDf(filepath)
```

```python
# Print only columns values in a dataframe
df=readDf(filepath)
def printColumn(df):
    columns=df.columns
    return columns
printColumn(df)
```

```python
df=readDf(filepath)
def printData(df):
    columns=df.columns
    for i in columns:
        print(i,end="  ")
    return
printData(df)

# Print row data
def printrowData(df):
    values=df.values
    for row in values:
        #print(row)
        for item in row:
            print(item,end="  ")
    return
printrowData(df)
```
