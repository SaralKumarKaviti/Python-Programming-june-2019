```python
# Linear graph

import matplotlib.pyplot as plt

plt.title('Basic Linear Graph')
plt.plot([1,2,3],[1,2,3])
plt.savefig('Images/Basic-Linear-Graph.png')
plt.show()
```

```python
# Graph to plot distance and time

distance=list(range(2,11,2))
time=list(range(1,6))

#distance
time
# d=2*t
plt.title('Distance vs Time')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time,distance,'ro')
plt.grid()
plt.savefig('Images/Distance-Time.png')
plt.show()

```

```python
# v= a t
t=list(range(1,11))
a=list(range(1,len(t)+2))
v=[a[i] * i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
plt.savefig('Images/Acceleration-Time.png')
plt.show()
```

```python
# Numbers vs squares
n=list(range(10))
sgn=[i ** 2for i in n]
plt.grid()
plt.plot(n,sgn)
plt.savefig('Images/Number-Squares.png')
plt.show()
```

```python
# Voltage vs Time
import numpy as np
t=[i/10 for i in range(1,21)]
v=[1+np.sin(2*np.pi*i) for i in t]
plt.title('Time vs Voltage')
plt.plot(t,v,'r')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Images/Voltage-Time.png')
plt.show()
```

```python
# Plot Income of Arakansas from 2005 t0 2013
import pandas as pd

df=pd.read_csv('DataFiles/Income.csv')
years=df.columns[2:]

income_Arkansas=df.values[3,2:]
plt.title('Income of Arkansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,income_Arkansas)
plt.savefig('Images/income_arkansas.png')
plt.show()
```
